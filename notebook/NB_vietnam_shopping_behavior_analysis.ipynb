```python
df['Review_Rating'] = df.groupby('Category')['Review_Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Teen', 'Young Adult', 'Adult']
df['age_group'] = pd.qcut(df['age'], q = 3, labels=labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
print(df['age_group'].value_counts())
print(df.groupby('age_group')['age'].agg(['min', 'max']))
```

```python
# create column purchase_frequency_days
# convert text to numeric for easier compare later 

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90,
}

df['purchase_frequency_days'] = df['purchase_frequency'].map(frequency_mapping)
```
